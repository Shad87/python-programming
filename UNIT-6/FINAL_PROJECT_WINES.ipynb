```python
winereviews = pd.read_csv('winemag-data1.csv')
```

```python
winereviews.shape
```

```python
winereviews.head()
```

```python
#Viewing columns in the Data Frame
winereviews.columns
```

```python
#Cleaning the data -- Capitalizing the Column names
winereviews.cols = ['Unnamed: 0','Country', 'Description', 'Designation', 'Points',
       'Price', 'Province', 'Region_1', 'Region_2', 'Taster_name',
       'Taster_twitter_handle', 'Title', 'Variety', 'Winery']
```

```python
winereviews.columns = winereviews.cols
```

```python
winereviews.head()
```

```python
#Cleaning the data -- dropping an unwanted column
winereviews.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
winereviews.head(60000)
```

```python
#Cleaning the data -- checking for missing values
winereviews.isnull().sum()
#winereviews.isnull().head()
#winereviews[winereviews.Region_2.isnull()]
```

```python
#Cleaning the data -- dropping multiple unwanted columns
winereviews.drop(['Region_2','Region_1', 'Taster_twitter_handle'], axis=1, inplace=True)
```

```python
winereviews.head()
```

```python
#Replacing null values -- Median Price for NaN 
winereviews['Price'].fillna(winereviews.groupby('Country')['Price'].transform("median"), inplace = True)
```

```python
winereviews.head()
```

```python
winereviews.isnull().sum()
```

```python
winereviews.shape
```

```python
#Dropping null str values
winereviews.dropna().shape
```

```python
winereviews.isnull().sum()
```

```python
#Dropping null str values -- 
winereviews.dropna(subset=['Country', 'Points', 'Price', 'Province', 
                           'Winery', 'Designation', 'Taster_name'], how = 'any', inplace = True)
```

```python
winereviews.isnull().sum()
```

```python
sbs.pairplot(winereviews) #exploring data relationships
```

```python
winereviews.describe()
```

```python
#THE MAX WINE RATING ACROSS ALL COUNTRIES 
winereviews.groupby('Country').Points.max().head(10)
```

```python
#THE MAX WINE RATING ACROSS ALL COUNTRIES
winereviews.groupby('Country')['Points'].max().plot('bar', color = (0.3,0.4,0.9,0.4))
fig = plt.gcf()
fig.set_size_inches(22,8) #sets the size of the chart
```

```python
#Showing the frequency of MAX Points rating occurenece in the data
winereviews.groupby('Country')['Points'].max().plot('hist', color = (0.1,0.5,0.3,0.2))
fig = plt.gcf()
fig.set_size_inches(12,5) #sets the size of the chart
```

```python
#list the top rated wine variety by points
winereviews.groupby(by='Variety')[['Points']].count().sort_values(by='Points', ascending=False).head()
#winereviewss['Variety'].value_counts().tail(5)
```

```python
#The countries with the highest TOTAL point ratings -- compare country to country on highest rated wines
winereviews.groupby(by='Country')[['Points']].count().sort_values(by='Points', ascending =False).head()
```

```python
winereviews.groupby(by ='Country')['Points'].count().plot('bar', color='r') #Countries with the highest TOTAL point ratings
fig = plt.gcf()
fig.set_size_inches(19,5) #sets the size of the chart
```

```python
#showing the country with the highest number of reviewed items
winereviews['Country'].value_counts().plot(color='g')
fig = plt.gcf()
fig.set_size_inches(9.5,9.5) #sets the size of the chart
```

```python
winereviews.groupby('Country')['Points'].mean().plot('hist',  width=0.70, figsize=(25,8.8), color = (0.5,0.1,0.5,0.6))
plt.title('Average Points by Country') #Labels the Chart
plt.xlabel('Country') #Labels the xaxis
plt.ylabel('Points') #Labels the yaxis
```
